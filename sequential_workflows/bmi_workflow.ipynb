```python
from langgraph.graph import StateGraph,START,END
from typing import TypedDict
```

```python
#defiine state

class BMIState(TypedDict):
    weight_kg:float
    height_m:float
    bmi:float
    bmi_label:str
    
     
```

```python
def label_bmi(state:BMIState)->BMIState:
    bmi = state['bmi']
    if bmi < 18.5:
        state['bmi_label'] = 'Underweight'
    elif bmi < 25:
        state['bmi_label'] = 'Normal weight'
    elif bmi < 30:
        state['bmi_label'] = 'Overweight'
    else:
        state['bmi_label'] = 'Obesity'
    return state
```

```python
def calculate_bmi(state:BMIState)->BMIState:
    state['bmi'] = round(state['weight_kg']/(state['height_m']**2),2)
    return state
```

```python
graph = StateGraph(BMIState)

#add nodes to your graph

graph.add_node("calculate_bmi",calculate_bmi)
graph.add_node("label_bmi",label_bmi)


#add edges to your graph

graph.add_edge(START,"calculate_bmi")
graph.add_edge("calculate_bmi","label_bmi")
graph.add_edge("label_bmi",END)



#complie the graph and run it

workflow = graph.compile()

```

```python

#execute the graph and get the result

output_state = workflow.invoke({'weight_kg':70,'height_m':1.75})
output_state
```
